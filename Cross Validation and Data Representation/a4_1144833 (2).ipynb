### Load the Dataset

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('adult.csv')
```

```python
import numpy as np
```

```python
replaced_data = data.replace(r'^\s*\?\s*$', np.nan, regex=True)
```

```python
rows_with_nan = replaced_data[replaced_data.isnull().any(axis=1)]
print("Total rows with missing values:", len(rows_with_nan))
```

```python
replaced_data.info()
```

### Drop Missing Value

```python
replaced_data = replaced_data.dropna()
```

```python
replaced_data.info()
```

```python
replaced_data
```

```python
print('dataset shape after removing missing value: ', replaced_data.shape)
```

### Convert Column

```python
replaced_data['income'] = (
    replaced_data['income'].astype(str).str.strip().str.replace('.', '', 
                                                                regex = False).map({'<=50K': 0, '>50K': 1}))
```

```python
print(replaced_data['income'].value_counts(dropna=False))
assert set(replaced_data['income'].dropna().unique()) == {0, 1}
```

### Data Representation

```python
from sklearn.preprocessing import LabelEncoder
```

```python
if 'gender' in replaced_data.columns:
    replaced_data = pd.get_dummies(replaced_data, columns=['gender'], drop_first=True)
for i in replaced_data.select_dtypes(include=['object']).columns:
    replaced_data[i] = LabelEncoder().fit_transform(replaced_data[i])
```

#### Separate Columns Target and Feature Columns

```python
Y = replaced_data.pop('income')
X = replaced_data

print("Feature shape:", X.shape)
print("Target shape:", Y.shape)
```

```python
X
```

```python
Y
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns

X[numeric_cols] = scaler.fit_transform(X[numeric_cols])

print("\nData after StandardScaler:")
print("Shape after scaling: ", X.shape)
print(X.head())
```

### Split Dataset

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42, stratify=Y
)
print("Training data shape :", X_train.shape, Y_train.shape)
print("Testing data shape  :", X_test.shape, Y_test.shape)
```

```python
skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=100)

for fold, (train_idx, val_idx) in enumerate(skf.split(X_train, Y_train), start=1):
    X_tr, X_val = X_train.iloc[train_idx], X_train.iloc[val_idx]
    Y_tr, Y_val = Y_train.iloc[train_idx], Y_train.iloc[val_idx]

    print(f"\nFold {fold}")
    print("Train fold shape      :", X_tr.shape, Y_tr.shape)
    print("Validation fold shape :", X_val.shape, Y_val.shape)
    print("\nX_train:")
    print(X_tr.head())
    print("\nY_train:")
    print(Y_tr.head())
    print("\nX_val:")
    print(X_val.head())
    print("\nY_val:")
    print(Y_val.head())
```
